```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import matplotlib.pyplot as plt

# Assuming the dataframe is already loaded and processed as df
# Sort the dataframe by timestamp
df = df.sort_values(by='timestamp')

# Prepare the data
features = df[['borrowRate', 'lenderRate', 'totalValueLockedUSD',
               'totalBorrowBalanceUSD', 'totalDepositBalanceUSD',
               'hourlyDepositUSD', 'hourlyRepayUSD',
               'hourlyBorrowUSD', 'hourlyWithdrawUSD']]
target = df['net_apy']
timestamps = df['timestamp']

# Split the data: Let's use the first 80% of the data for training and the last 20% for testing
split_index = int(0.8 * len(df))
X_train, X_test = features.iloc[:split_index], features.iloc[split_index:]
y_train, y_test = target.iloc[:split_index], target.iloc[split_index:]
timestamps_train, timestamps_test = timestamps.iloc[:split_index], timestamps.iloc[split_index:]

# Standardize the Data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build the Model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the Model
history = model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=1)

# Evaluate the Model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')

# Plot the Training and Validation Loss
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot the Training and Validation MAE
plt.figure(figsize=(10, 6))
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.show()

# Make Predictions
y_pred = model.predict(X_test)

# Convert timestamps to datetime for better plotting
timestamps_test = pd.to_datetime(timestamps_test, unit='s')

# Plot Actual vs Predicted Net APY with time axis
plt.figure(figsize=(14, 8))
plt.plot(timestamps_test, y_test, label='Actual Net APY', color='blue')
plt.plot(timestamps_test, y_pred, label='Predicted Net APY', color='red', linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Net APY')
plt.title('Actual vs Predicted Net APY Over Time')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from kerastuner import HyperModel
from kerastuner.tuners import RandomSearch
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming the dataframe is already loaded and processed as df
# Sort the dataframe by timestamp
df = df.sort_values(by='marketHourlySnapshots_timestamp')

# Prepare the data
features = df[['marketHourlySnapshots_exchangeRate', 'borrower_rate', 'lender_rate',
               'marketHourlySnapshots_totalValueLockedUSD', 'marketHourlySnapshots_totalBorrowBalanceUSD',
               'marketHourlySnapshots_totalDepositBalanceUSD', 'marketHourlySnapshots_hourlyDepositUSD',
               'marketHourlySnapshots_hourlyRepayUSD', 'marketHourlySnapshots_hourlyBorrowUSD',
               'marketHourlySnapshots_hourlyWithdrawUSD']]
target = df['net_apy']
timestamps = df['marketHourlySnapshots_timestamp']

# Split the data: Let's use the first 80% of the data for training and the last 20% for testing
split_index = int(0.8 * len(df))
X_train, X_test = features.iloc[:split_index], features.iloc[split_index:]
y_train, y_test = target.iloc[:split_index], target.iloc[split_index:]
timestamps_train, timestamps_test = timestamps.iloc[:split_index], timestamps.iloc[split_index:]

# Standardize the Data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the hypermodel
class RegressionHyperModel(HyperModel):
    def build(self, hp):
        model = tf.keras.Sequential()
        model.add(tf.keras.layers.Input(shape=(X_train.shape[1],)))

        # Tune the number of layers
        for i in range(hp.Int('num_layers', 1, 5)):
            model.add(tf.keras.layers.Dense(units=hp.Int(f'units_{i}', min_value=32, max_value=512, step=32),
                                            activation='relu'))

        model.add(tf.keras.layers.Dense(1))

        model.compile(optimizer=tf.keras.optimizers.Adam(hp.Choice('learning_rate', [1e-3, 1e-4, 1e-5])),
                      loss='mse',
                      metrics=['mae'])
        return model

# Initialize the tuner
tuner = RandomSearch(
    RegressionHyperModel(),
    objective='val_mae',
    max_trials=10,
    executions_per_trial=2,
    directory='my_dir',
    project_name='regression')

# Perform the hyperparameter search
tuner.search(X_train, y_train, epochs=100, validation_split=0.2)

# Get the optimal hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# Build the model with the optimal hyperparameters
model = tuner.hypermodel.build(best_hps)

# Train the model
history = model.fit(X_train, y_train, epochs=100, validation_split=0.2)

# Evaluate the Model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')

# Plot the Training and Validation Loss
fig_loss = go.Figure()
fig_loss.add_trace(go.Scatter(y=history.history['loss'], mode='lines', name='Training Loss'))
fig_loss.add_trace(go.Scatter(y=history.history['val_loss'], mode='lines', name='Validation Loss'))
fig_loss.update_layout(title='Training and Validation Loss', xaxis_title='Epoch', yaxis_title='Loss')
fig_loss.show()

# Plot the Training and Validation MAE
fig_mae = go.Figure()
fig_mae.add_trace(go.Scatter(y=history.history['mae'], mode='lines', name='Training MAE'))
fig_mae.add_trace(go.Scatter(y=history.history['val_mae'], mode='lines', name='Validation MAE'))
fig_mae.update_layout(title='Training and Validation MAE', xaxis_title='Epoch', yaxis_title='MAE')
fig_mae.show()

# Make Predictions
y_pred = model.predict(X_test)

# Convert timestamps to datetime for better plotting
timestamps_test = pd.to_datetime(timestamps_test, unit='s')

# Plot Actual vs Predicted Net APY with time axis
fig_actual_vs_predicted = go.Figure()
fig_actual_vs_predicted.add_trace(go.Scatter(x=timestamps_test, y=y_test, mode='lines', name='Actual Net APY'))
fig_actual_vs_predicted.add_trace(go.Scatter(x=timestamps_test, y=y_pred[:, 0], mode='lines', name='Predicted Net APY', line=dict(dash='dash')))
fig_actual_vs_predicted.update_layout(title='Actual vs Predicted Net APY Over Time', xaxis_title='Time', yaxis_title='Net APY')
fig_actual_vs_predicted.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def train_and_evaluate_model(df, feature_columns, target_column, timestamp_column='marketHourlySnapshots_timestamp'):
    # Sort the dataframe by timestamp
    df = df.sort_values(by=timestamp_column)

    # Prepare the data
    features = df[feature_columns]
    target = df[target_column]
    timestamps = df[timestamp_column]

    # Split the data: Let's use the first 80% of the data for training and the last 20% for testing
    split_index = int(0.8 * len(df))
    X_train, X_test = features.iloc[:split_index], features.iloc[split_index:]
    y_train, y_test = target.iloc[:split_index], target.iloc[split_index:]
    timestamps_train, timestamps_test = timestamps.iloc[:split_index], timestamps.iloc[split_index:]

    # Standardize the Data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Reshape data for GRU input
    X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    # Build the GRU model
    model = tf.keras.models.Sequential([
        tf.keras.layers.GRU(96, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
        tf.keras.layers.GRU(352, return_sequences=True),
        tf.keras.layers.GRU(96, return_sequences=True),
        tf.keras.layers.GRU(416, return_sequences=True),
        tf.keras.layers.GRU(512),
        tf.keras.layers.Dense(1)
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mae'])

    # Define early stopping
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Train the model
    history = model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=1, callbacks=[early_stopping])

    # Evaluate the model
    test_loss, test_mae = model.evaluate(X_test, y_test)
    print(f'Test MSE: {test_loss}, Test MAE: {test_mae}')

    # Make predictions
    y_pred = model.predict(X_test)

    # Compute the percentage error
    percentage_error = ((y_test.values - y_pred[:, 0]) / y_test.values) * 100

    # Convert timestamps to datetime for better plotting
    timestamps_test = pd.to_datetime(timestamps_test, unit='s')

    # Plot the Training and Validation Loss
    fig_loss = go.Figure()
    fig_loss.add_trace(go.Scatter(y=history.history['loss'], mode='lines', name='Training Loss'))
    fig_loss.add_trace(go.Scatter(y=history.history['val_loss'], mode='lines', name='Validation Loss'))
    fig_loss.update_layout(title=f'Training and Validation Loss (Test MSE: {test_loss:.4f})', xaxis_title='Epoch', yaxis_title='Loss')
    fig_loss.show()

    # Plot the Training and Validation MAE
    fig_mae = go.Figure()
    fig_mae.add_trace(go.Scatter(y=history.history['mae'], mode='lines', name='Training MAE'))
    fig_mae.add_trace(go.Scatter(y=history.history['val_mae'], mode='lines', name='Validation MAE'))
    fig_mae.update_layout(title='Training and Validation MAE', xaxis_title='Epoch', yaxis_title='MAE')
    fig_mae.show()

    # Plot Actual vs Predicted Net APY with time axis
    fig_actual_vs_predicted = go.Figure()
    fig_actual_vs_predicted.add_trace(go.Scatter(x=timestamps_test, y=y_test, mode='lines', name='Actual Net APY'))
    fig_actual_vs_predicted.add_trace(go.Scatter(x=timestamps_test, y=y_pred[:, 0], mode='lines', name='Predicted Net APY', line=dict(dash='dash')))
    fig_actual_vs_predicted.update_layout(title='Actual vs Predicted Net APY Over Time', xaxis_title='Time', yaxis_title='Net APY')
    fig_actual_vs_predicted.show()

    # Plot the percentage error with respect to time
    fig_percentage_error = go.Figure()
    fig_percentage_error.add_trace(go.Scatter(x=timestamps_test, y=percentage_error, mode='lines', name='Percentage Error'))
    fig_percentage_error.update_layout(title='Percentage Error of Predicted Net APY Over Time', xaxis_title='Time', yaxis_title='Percentage Error')
    fig_percentage_error.show()

# Example usage:
# Assuming you have a DataFrame named df with the necessary columns
feature_columns = ['marketHourlySnapshots_exchangeRate',
                   'marketHourlySnapshots_totalValueLockedUSD', 'marketHourlySnapshots_totalBorrowBalanceUSD',
                   'marketHourlySnapshots_totalDepositBalanceUSD', 'marketHourlySnapshots_hourlyDepositUSD',
                   'marketHourlySnapshots_hourlyRepayUSD', 'marketHourlySnapshots_hourlyBorrowUSD',
                   'marketHourlySnapshots_hourlyWithdrawUSD']
target_column = 'net_apy'

train_and_evaluate_model(df, feature_columns, target_column)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def train_and_predict_future(df, feature_columns, target_column, timestamp_column='marketHourlySnapshots_timestamp', future_steps=24):
    # Sort the dataframe by timestamp
    df = df.sort_values(by=timestamp_column)

    # Prepare the data
    features = df[feature_columns]
    target = df[target_column]
    timestamps = df[timestamp_column]

    # Split the data: Let's use the first 80% of the data for training and the last 20% for testing
    split_index = int(0.8 * len(df))
    X_train, X_test = features.iloc[:split_index], features.iloc[split_index:]
    y_train, y_test = target.iloc[:split_index], target.iloc[split_index:]
    timestamps_train, timestamps_test = timestamps.iloc[:split_index], timestamps.iloc[split_index:]

    # Standardize the Data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Reshape data for GRU input
    X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    # Build the GRU model
    model = tf.keras.models.Sequential([
        tf.keras.layers.GRU(96, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
        tf.keras.layers.GRU(352, return_sequences=True),
        tf.keras.layers.GRU(96, return_sequences=True),
        tf.keras.layers.GRU(416, return_sequences=True),
        tf.keras.layers.GRU(512),
        tf.keras.layers.Dense(1)
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mae'])

    # Define early stopping
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Train the model
    history = model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=1, callbacks=[early_stopping])

    # Evaluate the model
    test_loss, test_mae = model.evaluate(X_test, y_test)
    print(f'Test MSE: {test_loss}, Test MAE: {test_mae}')

    # Make future predictions
    future_predictions = []
    last_feature_set = X_test[-1].reshape(1, 1, -1)

    for _ in range(future_steps):
        prediction = model.predict(last_feature_set)
        future_predictions.append(prediction[0, 0])
        # Update the last_feature_set for the next prediction
        last_feature_set = np.roll(last_feature_set, -1)
        last_feature_set[0, 0, -1] = prediction

    # Convert timestamps to datetime for better plotting
    last_timestamp = pd.to_datetime(timestamps.iloc[-1], unit='s')
    future_timestamps = [last_timestamp + pd.Timedelta(hours=i) for i in range(1, future_steps + 1)]

    # Plot the Training and Validation Loss
    fig_loss = go.Figure()
    fig_loss.add_trace(go.Scatter(y=history.history['loss'], mode='lines', name='Training Loss'))
    fig_loss.add_trace(go.Scatter(y=history.history['val_loss'], mode='lines', name='Validation Loss'))
    fig_loss.update_layout(title='Training and Validation Loss', xaxis_title='Epoch', yaxis_title='Loss')
    fig_loss.show()

    # Plot the Training and Validation MAE
    fig_mae = go.Figure()
    fig_mae.add_trace(go.Scatter(y=history.history['mae'], mode='lines', name='Training MAE'))
    fig_mae.add_trace(go.Scatter(y=history.history['val_mae'], mode='lines', name='Validation MAE'))
    fig_mae.update_layout(title='Training and Validation MAE', xaxis_title='Epoch', yaxis_title='MAE')
    fig_mae.show()

    # Plot Actual vs Predicted Net APY with time axis
    fig_actual_vs_predicted = go.Figure()
    fig_actual_vs_predicted.add_trace(go.Scatter(x=pd.to_datetime(timestamps, unit='s'), y=target, mode='lines', name='Actual Net APY'))
    fig_actual_vs_predicted.add_trace(go.Scatter(x=future_timestamps, y=future_predictions, mode='lines', name='Predicted Net APY', line=dict(dash='dash')))
    fig_actual_vs_predicted.update_layout(title='Actual and Predicted Net APY Over Time', xaxis_title='Time', yaxis_title='Net APY')
    fig_actual_vs_predicted.show()

# Example usage:
# Assuming you have a DataFrame named df with the necessary columns
feature_columns = ['marketHourlySnapshots_exchangeRate', 'borrower_rate', 'lender_rate',
                   'marketHourlySnapshots_totalValueLockedUSD', 'marketHourlySnapshots_totalBorrowBalanceUSD',
                   'marketHourlySnapshots_totalDepositBalanceUSD', 'marketHourlySnapshots_hourlyDepositUSD',
                   'marketHourlySnapshots_hourlyRepayUSD', 'marketHourlySnapshots_hourlyBorrowUSD',
                   'marketHourlySnapshots_hourlyWithdrawUSD']
target_column = 'net_apy'

train_and_predict_future(df, feature_columns, target_column)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import shap

def create_model(input_shape):
    model = tf.keras.models.Sequential([
        tf.keras.layers.GRU(96, return_sequences=True, input_shape=input_shape),
        tf.keras.layers.GRU(352, return_sequences=True),
        tf.keras.layers.GRU(96, return_sequences=True),
        tf.keras.layers.GRU(416, return_sequences=True),
        tf.keras.layers.GRU(512),
        tf.keras.layers.Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mae'])
    return model

def custom_predict(model, data):
    data_reshaped = data.reshape((data.shape[0], 1, data.shape[1]))
    return model.predict(data_reshaped)

def train_and_predict_future(df, feature_columns, target_column, timestamp_column='marketHourlySnapshots_timestamp', future_steps=24):
    # Sort the dataframe by timestamp
    df = df.sort_values(by=timestamp_column)

    # Prepare the data
    features = df[feature_columns]
    target = df[target_column]
    timestamps = df[timestamp_column]

    # Split the data: Let's use the first 80% of the data for training and the last 20% for testing
    split_index = int(0.8 * len(df))
    X_train, X_test = features.iloc[:split_index], features.iloc[split_index:]
    y_train, y_test = target.iloc[:split_index], target.iloc[split_index:]
    timestamps_train, timestamps_test = timestamps.iloc[:split_index], timestamps.iloc[split_index:]

    # Standardize the Data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Build the GRU model
    model = create_model(input_shape=(1, X_train.shape[1]))

    # Reshape data for GRU input
    X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    # Define early stopping
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Train the model
    history = model.fit(X_train_reshaped, y_train, epochs=100, validation_split=0.2, verbose=1, callbacks=[early_stopping])

    # Evaluate the model
    test_loss, test_mae = model.evaluate(X_test_reshaped, y_test)
    print(f'Test MSE: {test_loss}, Test MAE: {test_mae}')


    # Make predictions on the training data for plotting
    train_predictions = model.predict(X_train_reshaped)

    # SHAP values calculation using KernelExplainer with custom prediction function
    explainer = shap.KernelExplainer(lambda x: custom_predict(model, x), X_train[:5])  # Use a sample for efficiency
    shap_values = explainer.shap_values(X_test[:5])

    # Plot SHAP values
    shap.summary_plot(shap_values, X_test[:5], feature_names=feature_columns)

    # Convert timestamps to datetime for better plotting
    last_timestamp = pd.to_datetime(timestamps.iloc[split_index - 1], unit='s')
    future_timestamps = [last_timestamp + pd.Timedelta(hours=i) for i in range(1, future_steps + 1)]

    # Plot the Training and Validation Loss
    fig_loss = go.Figure()
    fig_loss.add_trace(go.Scatter(y=history.history['loss'], mode='lines', name='Training Loss'))
    fig_loss.add_trace(go.Scatter(y=history.history['val_loss'], mode='lines', name='Validation Loss'))
    fig_loss.update_layout(title='Training and Validation Loss', xaxis_title='Epoch', yaxis_title='Loss')
    fig_loss.show()

    # Plot the Training and Validation MAE
    fig_mae = go.Figure()
    fig_mae.add_trace(go.Scatter(y=history.history['mae'], mode='lines', name='Training MAE'))
    fig_mae.add_trace(go.Scatter(y=history.history['val_mae'], mode='lines', name='Validation MAE'))
    fig_mae.update_layout(title='Training and Validation MAE', xaxis_title='Epoch', yaxis_title='MAE')
    fig_mae.show()

    # Plot Actual vs Predicted Net APY with time axis for training data
    fig_train_actual_vs_predicted = go.Figure()
    fig_train_actual_vs_predicted.add_trace(go.Scatter(x=pd.to_datetime(timestamps_train, unit='s'), y=y_train, mode='lines', name='Actual Train Net APY'))
    fig_train_actual_vs_predicted.add_trace(go.Scatter(x=pd.to_datetime(timestamps_train, unit='s'), y=train_predictions[:, 0], mode='lines', name='Predicted Train Net APY', line=dict(dash='dash')))
    fig_train_actual_vs_predicted.update_layout(title='Actual vs Predicted Train Net APY Over Time', xaxis_title='Time', yaxis_title='Net APY')
    fig_train_actual_vs_predicted.show()


# Example usage:
# Assuming you have a DataFrame named df with the necessary columns
feature_columns = ['marketHourlySnapshots_exchangeRate',
                   'marketHourlySnapshots_totalValueLockedUSD', 'marketHourlySnapshots_totalBorrowBalanceUSD',
                   'marketHourlySnapshots_totalDepositBalanceUSD', 'marketHourlySnapshots_hourlyDepositUSD',
                   'marketHourlySnapshots_hourlyRepayUSD', 'marketHourlySnapshots_hourlyBorrowUSD',
                   'marketHourlySnapshots_hourlyWithdrawUSD']
target_column = 'net_apy'

train_and_predict_future(df, feature_columns, target_column)
```
